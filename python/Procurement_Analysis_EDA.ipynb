Importing Dependencies

```python
import pandas as pd
```

Loading Dataset

```python
df = pd.read_csv("D:\DA\Portfolio Projects\Procurement\Procurement KPI Analysis Dataset.csv")
#print first 5 rows of the dataset
print(df.head())
```

Data Cleaning

```python
#Changing into Datetime
df['Order_Date'] = pd.to_datetime(df['Order_Date'])
df['Delivery_Date'] = pd.to_datetime(df['Delivery_Date'])
```

```python
#Check Nulls
df.isnull().sum()
```

```python
#Remove Duplicates
df.drop_duplicates(inplace=True)
```

```python
#Overview of Data
print(df.shape)
print(df.describe())
print(df.isnull().sum())
```

Feature Engineering

```python
#Fix Missing Values
df['Defective_Units'] = df['Defective_Units'].fillna(0)
```

```python
#Create a Delivery Flag
df['Delivery_Status'] = df['Delivery_Date'].apply(lambda x: "Delivered" if pd.notnull(x)else "Not Delivered")
```

```python
#Create Negotiative Savings
df['Negotiative_Savings'] = df['Unit_Price'] - df['Negotiated_Price']
```

```python
df.rename(columns={"Negotiative_Savings": "Negotiation_Savings"}, inplace=True)
```

```python
#Create Total Spend
df["Total_Cost"] = df["Negotiated_Price"] * df["Quantity"]
```

```python
#Defect Rate 
df['Defect_Rate'] = df['Defective_Units'] / df['Quantity']
```

```python
#Delivery Time
df['Delivery_Time'] = (df['Delivery_Date']-df['Order_Date']).dt.days
```

```python
#Compliance Status
df['Compliance_Status'] = df['Compliance'].map({"Yes":"Compliant","No":"Non-Compliant"})
```

# EDA

Top Vendors by Spend

```python
df.groupby("Supplier")['Total_Cost'].sum().sort_values(ascending=False)
```

Worst Vendors by Defect Rate

```python
df.groupby("Supplier")['Defect_Rate'].mean().sort_values(ascending=False)
```

Best Negotiators

```python
df.groupby("Supplier")['Negotiation_Savings'].mean().sort_values(ascending=False)
```

Compliance Check

```python
df.groupby("Supplier")['Compliance_Status'].value_counts()
```

Delta Logistics has most Non-Compliance Status

Cost vs Quality

```python
df.groupby("Supplier").agg({
            "Total_Cost":"sum",
            "Defect_Rate":"mean"
}).sort_values(by="Total_Cost",ascending=False)
```
